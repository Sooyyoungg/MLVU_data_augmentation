```python
import sys
import os
import xml.etree.ElementTree as Et
from xml.etree.ElementTree import Element, ElementTree

xml_path = '../../data/trainval/VOCdevkit/VOC2012/Annotations/'
py_dict = {}
print("XML parsing Start\n")
for file_name in file_list:
    xml = open(file_name, "r")
    tree = Et.parse(xml)
    root = tree.getroot()


   # size = root.find("size")
    segmented = root.find("segmented").text
#    width = size.find("width").text
 #   height = size.find("height").text
  #  channels = size.find("depth").text
    if segmented =='1':
        #print("segmented : {}\n".format(segmented))
        print(file_name)

        objects = root.findall("object")
        print("Objects Description")
        name_list= []
        for _object in objects:
            name = _object.find("name").text
            #bndbox = _object.find("bndbox")
            #xmin = bndbox.find("xmin").text
            #ymin = bndbox.find("ymin").text
            #xmax = bndbox.find("xmax").text
            #ymax = bndbox.find("ymax").text
            name_list.append(name)

            print("class : {}\n".format(name))
        py_dict[file_name] = name_list
#    print("XML parsing END")
```

```python
for key in py_dict.keys():
    #ind = 0
    
        
    segmentation_path = seg_path + key[-15:-4] + '.png'
    #segmentation_path = seg_path + key[-10:-4] + '.png'
    seg_image = Image.open(segmentation_path)
    array_pixel = np.array(seg_image)
    bb = 0
    for sss in py_dict[key]:
        aa = np.where(array_pixel == my_dict[sss], 1, 0)
        bb +=aa
    aa = np.clip(bb,0,1)
    image_path  = img_path + key[-15:-4] + '.jpg'
    image = Image.open(image_path)
    array_pixel = np.array(image)
        
    test = np.array(image)* np.stack([aa,aa,aa],axis =2)
    test1 = test.astype(np.uint8)
    im = Image.fromarray(test1)
    im.save('./black_data3/{}.jpg'.format(key[-15:-4]))  
```

```python
for key in py_dict.keys():
    #ind = 0
    
        
    segmentation_path = seg_path + key[-15:-4] + '.png'
    seg_image = Image.open(segmentation_path)
    array_pixel = np.array(seg_image)
    bb = 0
    for sss in py_dict[key]:
        aa = np.where(array_pixel == my_dict[sss], 1, 0)
        bb +=aa
    aa = np.clip(bb,0,1)
    image_path  = img_path + key[-15:-4] + '.jpg'
    image = Image.open(image_path)
    
    #ab = np.where(aa == 0, 255, 1)
    ac = np.zeros(aa.shape)
    gogo = np.where(aa == 1, -255, 0)
    gogo +=255
    
    array_pixel = np.array(image)     
    test = np.array(image)* np.stack([aa,aa,aa],axis =2)
    
    test = test + np.stack([gogo,ac,ac], axis =2)
    test = np.clip(test, 0,255)
    test1 = test.astype(np.uint8)
    im = Image.fromarray(test1)
    im.save('./red_data3/{}.jpg'.format(key[-15:-4]))  
```
